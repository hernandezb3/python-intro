## Booleans and comparisons

Control flow is built on conditions — expressions that evaluate to `True` or `False`.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | equal to | `5 == 5` | `True` |
| `!=` | not equal | `5 != 3` | `True` |
| `>` | greater than | `5 > 3` | `True` |
| `<` | less than | `5 < 3` | `False` |
| `>=` | greater or equal | `5 >= 5` | `True` |
| `<=` | less or equal | `3 <= 5` | `True` |
| `in` | membership | `'a' in 'cat'` | `True` |
| `not` | negate | `not True` | `False` |

````python
# comparison operators
x = 10
print(x == 10)   # True
print(x == 11)   # False
print(x > 5)     # True
print(x != 10)   # False

# combining conditions with 'and' / 'or'
age = 25
has_id = True

print(age >= 18 and has_id)    # both must be True
print(age < 18 or has_id)      # at least one must be True
print(not has_id)              # flips True to False
````

## if / elif / else

An `if` block runs its code only when the condition is `True`.
`elif` (else if) checks additional conditions. `else` catches everything that didn't match.

**Indentation is not optional.** Python uses 4 spaces to define which code belongs inside a block.
Inconsistent indentation is flagged as a syntax error.

````python
temperature = 72

if temperature > 85:
    print("Hot day")
elif temperature > 65:
    print("Nice day")    # this runs
elif temperature > 45:
    print("Cool day")
else:
    print("Cold day")

# only ONE branch ever runs — Python stops checking once it finds a match
````

````python
# nested if: a condition inside another condition
score = 83
submitted_on_time = True

if score >= 60:
    if submitted_on_time:
        print("Passed — full credit")
    else:
        print("Passed — but late penalty applies")
else:
    print("Did not pass")

# note: deeply nested ifs can be a sign to write a function instead
````

````python
# membership checks with 'in'
valid_colors = ["red", "green", "blue"]
user_input = "green"

if user_input in valid_colors:
    print(f"{user_input} is a valid color")
else:
    print(f"{user_input} is not recognized")
````

## Intro to debugging

Before we get into loops and functions — where bugs get more interesting — it's worth building
a systematic habit now.

### The debugging mindset
Good debugging is forming a hypothesis about what the code *is* doing
vs. what you *expect* it to do, then testing that hypothesis.

**Step-by-step process:**
1. **Read the error message** — what type is it? what line?
2. **Reproduce it** — can you make it fail consistently?
3. **Isolate it** — comment out code until you find the smallest piece that breaks
4. **Inspect state** — print the variables at the point of failure
5. **Fix and verify** — confirm the fix works, and didn't break something else

### How errors appear in functions:

| Error | Cause | Example |
|-------|-------|---------|
| `IndentationError` | wrong number of spaces inside a block | missing 4 spaces after `if` |
| `SyntaxError` | missing colon after `if`/`for`/`def` | `if x > 5` (no `:`) |
| `TypeError` | comparing incompatible types | `"5" > 3` |
| `NameError` | calling a function before defining it | `greet()` before `def greet` |
| `UnboundLocalError` | using a variable inside a function before assigning it | scope issue |

````python
# the single most useful debugging tool: print()
# when your logic isn't doing what you expect, print the values at each step

score = 73
print(f"score is: {score}, type is: {type(score)}")

if score >= 70:
    print("Checking: entered the >= 70 branch")
    grade = "C"
    print(f"grade assigned: {grade}")

# trace prints like these are temporary — remove them once the bug is fixed
````

## for loops

A `for` loop repeats a block of code once for **each item** in a sequence.
The loop variable takes on each value in turn — you name it, and it only exists inside the loop.

````python
# loop over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

# loop over a string (each character)
for char in "hello":
    print(char, end=" ")
print()   # newline after
````

````python
# range() generates a sequence of numbers
# range(stop)          — 0 up to (not including) stop
# range(start, stop)   — start up to (not including) stop
# range(start, stop, step)

for i in range(5):
    print(i, end=" ")       # 0 1 2 3 4
print()

for i in range(2, 10, 2):
    print(i, end=" ")       # 2 4 6 8
print()

for i in range(5, 0, -1):
    print(i, end=" ")       # 5 4 3 2 1
print()
````

````python
# enumerate() gives you the index AND the value — use this instead of range(len(...))
students = ["Alice", "Bob", "Carol"]

for i in range(len(students)):
    print(i, students[i])

print()

# instead you can use:
for i, student in enumerate(students):
    print(f"{i}: {student}")
````

````python
# break exits the loop early
# continue skips to the next iteration

numbers = [3, 7, 2, 9, 1, 5]

# stop as soon as we find a number > 8
for n in numbers:
    if n > 8:
        print(f"Found it: {n}")
        break

# skip even numbers
for n in numbers:
    if n % 2 == 0:
        continue
    print(n, end=" ")    # only odd numbers
````

## while loops

A `while` loop keeps running as long as its condition is `True`.
Use it when you don't know in advance how many iterations you need.

> **Debugging warning:** a `while` loop with a condition that never becomes `False` runs forever.
> If your cell seems frozen, go to Kernel → Interrupt.

````python
# count down from 5
count = 5
while count > 0:
    print(count, end=" ")
    count -= 1      # count = count - 1
print("Go!")
````

````python
# while is natural for 'keep asking until valid input'
# (in a notebook, input() pauses for you to type)

# simulating it with a list instead of live input:
responses = ["maybe", "yes"]    # pretend these are typed one at a time
idx = 0

answer = responses[idx]
while answer != "yes":
    print(f"Got '{answer}' — not valid, trying again")
    idx += 1
    answer = responses[idx]

print("Valid answer received!")
````

````python
# infinite loop
# the variable that the condition depends on MUST change inside the loop

# don't run this — it never stops:
# x = 0
# while x < 5:
#     print(x)      # x never changes — infinite loop!

# instead:
x = 0
while x < 5:
    print(x, end=" ")
    x += 1          # condition variable must update
````

## Functions

A function is a named, reusable block of code.

**Why functions matter:**
- Avoid copy-pasting the same logic in multiple places
- Give a name to a concept, making code self-documenting
- Make code testable — you can check one function in isolation

```python
def function_name(parameter1, parameter2):
    """Docstring: describe what this function does."""
    # body
    return result
```

PEP 8: function names use `snake_case`. Always include a docstring.

````python
# simplest function: no parameters, no return value
def greet():
    """Print a greeting."""
    print("Hello!")

greet()     # calling the function
greet()     # call it as many times as you like
````

````python
# function with parameters
def greet_person(name, time_of_day):
    """Print a personalised greeting."""
    print(f"Good {time_of_day}, {name}!")

greet_person("Alice", "morning")
greet_person("Bob", "evening")

# keyword arguments — order doesn't matter when you name them
greet_person(time_of_day="afternoon", name="Carol")
````

### Return values

`return` sends a value back to the caller. Without it, a function returns `None`.
A function exits immediately when it hits `return` — anything after it in the same branch won't run.

````python
def square(n):
    """Return the square of n."""
    return n ** 2

result = square(5)
print(result)                  # 25
print(square(3) + square(4))   # use the return value in an expression

# what happens without return?
def square_no_return(n):
    n ** 2    # computed but thrown away

print(square_no_return(5))     # None — a very common mistake
````

````python
# early return: exit as soon as you have the answer
def letter_grade(score):
    """Return a letter grade for a numeric score."""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(letter_grade(95))
print(letter_grade(72))
print(letter_grade(55))

# Now reuse it in a loop
scores = [91, 74, 58, 85, 63]
for score in scores:
    print(f"{score} is a {letter_grade(score)}")
````

### Default and keyword arguments

**Default arguments** give a parameter a fallback value when the caller doesn't provide one.
Put them *after* required parameters in the definition.

> **Debugging warning:** don't use a mutable object (list, dict) as a default argument value —
> it's shared across all calls of a function and causes subtle bugs. Use `None` and create the object inside the function instead.

````python
def make_label(text, color = "black", bold = False):
    """Return a formatted label string."""
    prefix = "**" if bold else ""
    return f"[{color.upper()}] {prefix}{text}{prefix}"

print(make_label("Warning"))                        # uses defaults
print(make_label("Error", color = "red"))             # override color
print(make_label("Critical", color = "red", bold = True))  # override both
````

````python
# the mutable default argument trap

# BAD — the list is created once and shared across all calls:
def add_item_bad(item, collection = []):
    collection.append(item)
    return collection

print(add_item_bad("apple"))    # ['apple']
print(add_item_bad("banana"))   # ['apple', 'banana'] — surprise!

# GOOD — create a fresh list each call:
def add_item_good(item, collection = None):
    if collection is None:
        collection = []
    collection.append(item)
    return collection

print(add_item_good("apple"))   # ['apple']
print(add_item_good("banana"))  # ['banana']
````

### Scope

**Scope** is where a variable can be seen and used.

- **Local scope:** variables created inside a function. They only exist while the function runs.
- **Global scope:** variables created at the top level of a script. Visible everywhere.

The rule: functions can *read* globals, but they can't *modify* them without the `global` keyword
(which you typically don't want to use — passing values as arguments is cleaner).

````python
# local variable: only visible inside the function
def calculate_tax(price):
    rate = 0.08          # local — only exists here
    return price * rate

print(calculate_tax(100))
# print(rate)   # NameError — 'rate' doesn't exist out here
````

````python
# functions can read global variables (but this is often bad practice)
discount = 0.10    # global

def apply_discount(price):
    """Apply the global discount rate."""
    return price * (1 - discount)   # reads global

print(apply_discount(50))

# Better: pass the value explicitly as an argument
def apply_discount_explicit(price, rate):
    """Apply a given discount rate."""
    return price * (1 - rate)

print(apply_discount_explicit(50, 0.10))
````

````python
# UnboundLocalError — the scope trap
# python sees an assignment to 'total' inside the function and treats it as local.
# but we try to read it before assigning — error.

total = 100

def broken_function():
    # total = total + 10    # UnboundLocalError: local var 'total' referenced before assignment
    pass

# Fix: pass total as a parameter
def fixed_function(total):
    total = total + 10
    return total

print(fixed_function(100))
````

## Putting it all together: loops + functions

Functions and loops are most powerful in combination.
A clean pattern: write a function that handles *one item*, then loop over a collection calling it.

````python
def classify_temperature(temp_f):
    """Return a string category for a temperature in Fahrenheit."""
    if temp_f >= 90:
        return "hot"
    elif temp_f >= 70:
        return "warm"
    elif temp_f >= 50:
        return "cool"
    else:
        return "cold"

daily_highs = [88, 72, 65, 94, 55, 78, 91]

for i, temp in enumerate(daily_highs):
    category = classify_temperature(temp)
    print(f"Day {i+1}: {temp}°F — {category}")
````
