## Version control with Git

> **These commands run in your terminal, not in notebook cells.**

Git tracks changes to your code over time. Think of it as a detailed save history —
you can always go back to any earlier version, see exactly what changed and when,
and collaborate with others without overwriting each other's work.

### The three-stage model

```
Working directory  →  Staging area  →  Repository
  (your files)         (git add)        (git commit)
```

- **Working directory:** your files as they are on disk right now
- **Staging area:** changes you've selected to include in the next commit
- **Repository:** the permanent history of committed snapshots

### Setting up a repo

```bash
# One-time global config (do this once on any new machine)
git config --global user.name  "Your Name"
git config --global user.email "you@example.com"

# Initialise a new repo in your project folder
cd my_project
git init
```

### The daily workflow

```bash
git status                  # what has changed?
git add filename.py         # stage one file
git add .                   # stage everything changed
git commit -m "message"     # save a snapshot with a description
git log --oneline           # see the history
```

### Writing good commit messages

A commit message should complete the sentence *"This commit will..."*

| ❌ Avoid | ✅ Prefer |
|---------|----------|
| `stuff` | `add grade calculator function` |
| `fix` | `fix off-by-one error in score loop` |
| `wip` | `add CSV export for student records` |
| `asdfg` | `rename variable for PEP 8 compliance` |

### What goes in .gitignore

Create a `.gitignore` file in your repo root. Git will never track these:

```
# Python
__pycache__/
*.pyc
.ipynb_checkpoints/

# Virtual environment
venv/
.env

# OS files
.DS_Store
Thumbs.db

# Data files you don't want to commit
*.csv
```

## Branching, merging & pull requests

> **Terminal commands — not notebook cells.**

A **branch** is an independent line of development. You make changes on a branch
without touching `main` — then merge when the work is ready.

This is the foundation of every professional code review workflow.

### Why branch?
- Experiment safely — `main` stays clean and working
- Work on multiple features in parallel
- Get feedback via a pull request before merging

### Branch workflow

```bash
# See all branches (* marks the current one)
git branch

# Create and switch to a new branch
git checkout -b feature/add-csv-export

# ... make changes, then stage and commit as normal ...
git add .
git commit -m "add CSV export for student records"

# Switch back to main
git checkout main

# Merge your branch into main
git merge feature/add-csv-export

# Delete the branch once merged (keep things tidy)
git branch -d feature/add-csv-export
```

### Branch naming conventions

| Type | Pattern | Example |
|------|---------|---------|
| Feature | `feature/short-description` | `feature/add-csv-export` |
| Bug fix | `fix/short-description` | `fix/off-by-one-in-loop` |
| Docs | `docs/short-description` | `docs/update-readme` |

### Connecting to GitHub and pushing

```bash
# Link your local repo to a GitHub repo (do once)
git remote add origin https://github.com/yourname/your-repo.git

# Push main for the first time
git push -u origin main

# Push a feature branch
git push -u origin feature/add-csv-export
```

### Opening a pull request

After pushing a branch to GitHub:
1. Go to your repo on github.com — GitHub shows a banner: *"Compare & pull request"*
2. Click it, write a description of what changed and why
3. Request a reviewer if working with others
4. Once approved (or self-reviewed), click **Merge pull request**
5. Delete the branch on GitHub
6. Pull the updated `main` back locally: `git pull origin main`

### Merge conflicts

A conflict happens when two branches changed the same line. Git marks the file:

```
<<<<<<< HEAD
    return "C"    ← your version
=======
    return "D"    ← incoming version
>>>>>>> feature/grade-fix
```

To resolve: edit the file to keep what you want (remove all the `<<<<`, `====`, `>>>>` markers),
then `git add` the file and `git commit`.

Conflicts sound scary but are routine — every developer hits them regularly.
