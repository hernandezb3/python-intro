### Arithmetic

Python supports standard math operations for string and integers, plus a couple of useful extras.

| Operator | Meaning | Example | Result |

### String operations
You can index a string from it's start, using positive values, or from the end, using negative values. 

From the start, indexing starts at `0`. From the end, indexing starts at `-1`. For example, in the code below, you can index the "t" in "Python" both using `2` or `-4`.

````python
word = "Python"

# length
print(len(word))          # 6

# indexing — counting starts at 0
print(word[0])            # P  (first character)
print(word[-1])           # n  (last character)

# slicing [start:stop] — stop is exclusive
print(word[0:3])          # Pyt
print(word[2:])           # thon (from index 2 to end)
print(word[:4])           # Pyth (from start to index 4)
````

````python
print(word[2])
print(word[-4])
````

````python
# useful string operations
messy = "  Hello, World!  "

print(messy.strip())       # remove leading/trailing whitespace
print(messy.lower())       # all lowercase
print(messy.upper())       # all uppercase
print(messy.replace("World", "Python"))   # find and replace
print(messy.strip().split(","))           # split into a list
````

### f-strings (formatted string literals)

f-strings are the cleanest way to build strings that include variable values.
Put an `f` before the opening quote, then use `{}` to embed any expression.

````python
name = "Marie"
age = 34
city = "Paris"

# alternative way
print("My name is " + name + " and I am " + str(age) + " years old.")

# f-string way
print(f"My name is {name} and I am {age} years old.")

# you can put expressions inside the braces
price = 9.99
quantity = 3
print(f"Total: ${price * quantity:.2f}")  # :.2f = 2 decimal places
````

### Lists

A **list** is an ordered collection of values. Think of it as a single variable that holds multiple items.

- Create lists with square brackets: `[item1, item2, item3]`
- Items can be any type, and you can mix types
- Lists are *ordered* — items keep the position you give them
- Lists are *mutable* — you can add, remove, and change items after creation

````python
# creating lists
fruits = ["apple", "banana", "cherry", "date"]
temperatures = [72.1, 68.4, 75.0, 80.2, 69.9]
mixed = ["Alice", 30, True, 5.5]        # valid, but unusual
empty = []                              # an empty list

print(fruits)
print(temperatures)
print(f"Number of fruits: {len(fruits)}")
````

#### Indexing and slicing

Lists use the same indexing rules as strings — counting starts at 0, and negative numbers count from the end.

````python
fruits = ["apple", "banana", "cherry", "date"]

# indexing
print(fruits[0])    # apple  — first item
print(fruits[2])    # cherry — third item
print(fruits[-1])   # date   — last item
print(fruits[-2])   # cherry — second to last

# slicing [start:stop] — stop index is excluded
print(fruits[1:3])  # ['banana', 'cherry']
print(fruits[:2])   # ['apple', 'banana']
print(fruits[2:])   # ['cherry', 'date']
````

#### Modifying lists

Unlike strings, lists are mutable — you can change them in place.

````python
fruits = ["apple", "banana", "cherry"]

# change an item
fruits[1] = "blueberry"
print(fruits)                   # ['apple', 'blueberry', 'cherry']

# add items
fruits.append("date")           # add to the end
fruits.insert(1, "avocado")     # insert at position 1
print(fruits)

# remove items
fruits.remove("cherry")         # remove by value
popped = fruits.pop()           # remove and return the last item
print(fruits)
print(f"Popped: {popped}")
````

#### Useful list operations

````python
scores = [88, 72, 95, 61, 84, 90, 78]

print(len(scores))          # number of items
print(min(scores))          # lowest value
print(max(scores))          # highest value
print(sum(scores))          # total
print(sum(scores)/len(scores))  # average

# sorting
print(sorted(scores))           # returns a new sorted list
scores.sort(reverse=True)       # sorts in place, descending
print(scores)

# check membership
print(95 in scores)             # True
print(100 in scores)            # False
````

### Tuples

A tuple is an **ordered, immutable** sequence — like a list that can never be changed after creation.

| | List | Tuple |
|--|------|-------|
| Syntax | `[1, 2, 3]` | `(1, 2, 3)` |
| Ordered | ✅ | ✅ |
| Mutable | ✅ | ❌ |
| Use when | data will change | data are fixed |

**When to use a tuple:**
- Coordinates, RGB colours, lat/long, names/ids — things that are naturally grouped and shouldn't change
- Returning multiple values from a function
- Dictionary keys (lists can't be keys; tuples can)

````python
# creating tuples
point = (3, 7)
rgb = (255, 128, 0)
single = (42,)       # trailing comma required for single-item tuple
also_tuple = 1, 2, 3 # parentheses optional — the comma makes it a tuple

print(point)
print(type(point))
print(len(rgb))
````

````python
# indexing and slicing work the same as lists
rgb = (255, 128, 0)
print(rgb[0])      # 255
print(rgb[-1])     # 0
print(rgb[:2])     # (255, 128)

# tuples are immutable — this raises a TypeError:
# rgb[0] = 100     # TypeError: 'tuple' object does not support item assignment
````

````python
# tuple unpacking — assign each element to a variable in one line
point = (3, 7)
x, y = point
print(f"x={x}, y={y}")

# swap two variables without a temp variable
a, b = 10, 20
a, b = b, a
print(f"a={a}, b={b}")    # a=20, b=10

# functions that return multiple values use tuples
def min_max(numbers):
    """Return the min and max of a list."""
    return min(numbers), max(numbers)   # returns a tuple

low, high = min_max([4, 1, 9, 2, 7])
print(f"Low: {low}, High: {high}")
````

### Sets

A set is an **unordered collection of unique values**. Duplicates are automatically removed.

| | List | Set |
|--|------|-----|
| Syntax | `[1, 2, 2]` | `{1, 2, 2}` → `{1, 2}` |
| Ordered | ✅ | ❌ |
| Duplicates | ✅ | ❌ |
| Fast membership test | slow (`O(n)`) | fast (`O(1)`) |

**When to use a set:**
- Deduplicating a list
- Checking membership when you don't need order
- Set operations: union, intersection, difference

````python
# creating sets
colours = {"red", "green", "blue", "red"}   # duplicate dropped silently
print(colours)       # order is not guaranteed
print(type(colours))

# create a set from a list — the fastest way to deduplicate
tags = ["python", "data", "python", "csv", "data", "pandas"]
unique_tags = set(tags)
print(unique_tags)
print(f"Original: {len(tags)} items, Unique: {len(unique_tags)}")
````

````python
# membership test — faster than 'in' on a list for large collections
valid_statuses = {"active", "pending", "closed"}

status = "pending"
print(status in valid_statuses)      # True
print("deleted" in valid_statuses)   # False

# adding and removing
valid_statuses.add("archived")
valid_statuses.discard("closed")     # discard won't error if item missing
print(valid_statuses)
````

````python
# set operations — useful for comparing datasets
week1_visitors = {"Amelia", "Ada", "Austin", "Aaron", "Bob"}
week2_visitors = {"Bob", "Carol", "Chloe", "Charlie", "Calvin"}

print(week1_visitors | week2_visitors)   # union: in either set
print(week1_visitors & week2_visitors)   # intersection: in both sets
print(week1_visitors - week2_visitors)   # difference: in week1 set only (not both sets or week2 set)
print(week1_visitors ^ week2_visitors)   # symmetric difference: unique items from each set
````

### Dictionaries

A **dictionary** stores data as **key-value pairs**. Instead of looking up items by position (index), you look them up with a meaningful name (key).

Think of it like a real dictionary: you look up a word (key) to find its definition (value).

- Create dictionaries with curly braces: `{key: value, key: value}`
- Keys are usually strings; values can be anything
- *Unordered* by default (though Python 3.7+ preserves insertion order)
- *Mutable* — you can add, change, and remove pairs

> **Why dicts matter for data work:** every row of a CSV is essentially a dictionary — column name mapped to cell value. Pandas DataFrames are built on this idea.

````python
# Creating a dictionary
person = {
    "name": "Ada Lovelace",
    "age": 36,
    "city": "London",
    "is_famous": True
}

print(person)
print(type(person))
````

#### Accessing values

````python
person = {"name": "Ada Lovelace", "age": 36, "city": "London"}

# Access by key
print(person["name"])        # Ada Lovelace
print(person["age"])         # 36

# .get() is safer — returns None instead of an error if the key doesn't exist
print(person.get("city"))    # London
print(person.get("email"))   # None (no error)
print(person.get("email", "no email on file"))  # custom default

# What happens with a missing key using []? Uncomment to see:
# print(person["email"])      # KeyError!
````

#### Adding and modifying entries

````python
person = {"name": "Ada Lovelace", "age": 36}

# add a new key-value pair
person["email"] = "ada@lovelace.com"
print(person)

# update an existing value
person["age"] = 37
print(person)

# remove a key
del person["email"]
print(person)

# remove and return a value
age = person.pop("age")
print(f"Removed age: {age}")
print(person)
````

## Reading errors (debugging basics)

Python errors always tell you *what* went wrong and *where*. The key is learning to read them.

The three most common errors in week 1:

````python
# 1. SyntaxError — you wrote something Python can't parse

# uncomment and run to see the error:
print("Hello")   # missing closing parenthesis
````

````python
# 2. NameError — you used a variable that doesn't exist yet

# uncomment and run to see the error:
# print(messsage)   # typo in variable name
````

````python
# 3. TypeError — you mixed incompatible types

# uncomment and run to see the error:
# age = 25
# print("I am " + age + " years old.")   # can't concatenate str and int
````

**How to read a traceback:**
1. Go to the **last line** first — that's the error type and message
2. Look at the **line number** just above it — that's where Python got confused
3. Read the **arrow** pointing at the specific part of the line

The error message is Python trying to help you. Read it literally before searching online.
